```python
tdhst = fits.open('3dhst.v4.1.5.master.fits.gz')
```

```python
data_dict= dict()
for obj in tdhst[1].data:
    for key in ['grism_id' , 'jh_mag']:
        data_dict.setdefault(key, list()).append(obj[key])
tddf = pd.DataFrame(data_dict)
```

```python
hdul = fits.open('grizli_v2.1.fits')
```

```python
hdul[1].header
```

```python
data_dict= dict()
for obj in hdul[1].data:
    for key in ['FIELD' , 'ID' , 'Z_50' , 'HA_FLUX' , 'HA_FLUX_ERR' , 'HB_FLUX' , 'HB_FLUX_ERR' , 'OII_FLUX' , 'OII_FLUX_ERR' ,'OIII_FLUX' , 'OIII_FLUX_ERR' , 'NEIII_3867_FLUX' , 'NEIII_3867_FLUX_ERR' ]:
        data_dict.setdefault(key, list()).append(obj[key])
cleardf = pd.DataFrame(data_dict)
```

```python
halphadf = cleardf[cleardf.Z_50 > .7]
halphadf = halphadf[halphadf.Z_50 < 1.6]
halphadf = halphadf[halphadf.HA_FLUX >= 0]
halphadf = halphadf[halphadf.HA_FLUX_ERR >= 0]
halphadf = halphadf[halphadf.HA_FLUX/halphadf.HA_FLUX_ERR >= 3]
halphadf = halphadf[halphadf.HB_FLUX >= 0]
halphadf = halphadf[halphadf.HB_FLUX_ERR >= 0]
halphadf = halphadf[halphadf.HB_FLUX/halphadf.HB_FLUX_ERR >= 3]
display(halphadf)
```

```python
o3df = cleardf[cleardf.Z_50 > 1.2]
o3df = o3df[o3df.Z_50 < 2.4]
o3df = o3df[o3df.OIII_FLUX >= 0]
o3df = o3df[o3df.OIII_FLUX_ERR >= 0]
o3df = o3df[o3df.OIII_FLUX/o3df.OIII_FLUX_ERR >= 3]
o3df = o3df[o3df.OII_FLUX >= 0]
o3df = o3df[o3df.OII_FLUX_ERR >= 0]
o3df = o3df[o3df.OII_FLUX/o3df.OII_FLUX_ERR >= 3]
o3df = o3df[o3df.HB_FLUX >= 0]
o3df = o3df[o3df.HB_FLUX_ERR >= 0]
o3df = o3df[o3df.HB_FLUX/o3df.HB_FLUX_ERR >= 3]
display(o3df)
```

```python
agndf = cleardf[cleardf.Z_50 > 1.2]
agndf = agndf[agndf.Z_50 < 2.4]
agndf = agndf[agndf.NEIII_3867_FLUX > 0]
agndf = agndf[agndf.NEIII_3867_FLUX_ERR > 0]
agndf = agndf[agndf.NEIII_3867_FLUX/agndf.NEIII_3867_FLUX_ERR > 3]
agndf = agndf[agndf.OII_FLUX > 0]
agndf = agndf[agndf.OII_FLUX_ERR > 0]
agndf = agndf[agndf.OII_FLUX/agndf.OII_FLUX_ERR > 3]
agndf = agndf[agndf.OIII_FLUX > 0]
agndf = agndf[agndf.OIII_FLUX_ERR > 0]
agndf = agndf[agndf.OIII_FLUX/agndf.OIII_FLUX_ERR > 3]
agndf = agndf[agndf.HB_FLUX > 0]
agndf = agndf[agndf.HB_FLUX_ERR > 0]
agndf = agndf[agndf.HB_FLUX/agndf.HB_FLUX_ERR > 3]
display(agndf)
print(len(agndf))
```
